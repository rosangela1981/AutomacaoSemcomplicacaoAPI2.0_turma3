```python
import numpy as np
import pandas as pd
df_salarios = pd.DataFrame({
   'nome' : ["Ana", "Bruno", "Carlos", "Daniele", "Val"],
   'salario' : [4000, np.nan, 5000, np.nan, 100000]
})

df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    "Dia" : ["segunda", "Terça", "Quarta", "Quinta", "sexta"],
    "Temperatura": [30, np.nan, 28,27, np.nan] # Added np.nan to match the length of 'Dia'
})

df_temperaturas["Preenchido_ffill"] = df_temperaturas["Temperatura"].ffill()
df_temperaturas
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df_limpo, x='coluna3', y='coluna1') # Alterado 'usd' para 'coluna1'
plt.title("Salário médio por senioridade") # O título pode precisar ser ajustado se 'senioridade' não for uma coluna ou conceito que você quer plotar com 'coluna3' e 'coluna1'
plt.xlabel ("coluna3") # Ajustado o rótulo do eixo x
plt.ylabel("Valor da Coluna 1") # Ajustado o rótulo do eixo y
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_limpo, x='coluna3', y='coluna1')
plt.title('Gráfico de Dispersão: Coluna3 vs Coluna1')
plt.xlabel('Coluna 3')
plt.ylabel('Coluna 1')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
